```python
# cleaning the city column of geolocation_df
from unidecode import unidecode
from collections import Counter

# Load (do this once)
geolocation_df = pd.read_csv("../data/olist_geolocation_dataset.csv")

# Step 1: Full standardization (run even if you already did unidecode)
geolocation_df['city_raw'] = geolocation_df['geolocation_city'].astype(str)
geolocation_df['city_clean'] = (geolocation_df['city_raw']
                        .apply(unidecode)
                        .str.lower()
                        .str.strip()
                            .str.replace(r'\s+', ' ', regex=True)   # collapse multiple spaces
                            .str.replace(r'[^a-z0-9\s]', '', regex=True))  # optional: remove any remaining punctuation

print("Unique cities before standardization:", geolocation_df['city_raw'].nunique())
print("Unique cities after basic clean:", geolocation_df['city_clean'].nunique())

# Step 2: Resolve per zip using mode (most common spelling wins)
# This fixes the real typos/variations automatically
city_map = (geolocation_df.groupby('geolocation_zip_code_prefix')['city_clean']
            .agg(lambda x: Counter(x).most_common(1)[0][0] if len(x) > 0 else x.iloc[0]))

geolocation_df['geolocation_city_clean'] = geolocation_df['geolocation_zip_code_prefix'].map(city_map)

# Optional: drop the old columns if you want a clean final df
geolocation_df = geolocation_df.drop(columns=['geolocation_city', 'city_raw', 'city_clean'])
geolocation_df = geolocation_df.rename(columns={'geolocation_city_clean': 'geolocation_city'})

print("Unique cities AFTER per-zip standardization:", geolocation_df['geolocation_city'].nunique())

# removing exact duplicates
# geolocation_df = geolocation_df.drop_duplicates()

# Checking the number of unique values in the city column
# geolocation_df['geolocation_city'].nunique()
```

```python
# Standardizing both dataset's city names

import unidecode
import re

def clean_column(df, column_name):
    
    df[column_name] = (
        df[column_name]
        .str.lower()                      # lowercase
        .str.strip()                      # remove extra spaces
        .apply(unidecode.unidecode)       # remove accents
        .str.replace("'", "", regex=False) # remove apostrophes
        .str.replace("-", " ", regex=False) # replace hyphen with space
        .str.replace(r"\s+", " ", regex=True) # remove double spaces
    )
    
    return df   

customers = clean_column(customers_df, 'customer_city')
geo = clean_column(geolocation_df, 'geolocation_city')
```

```python
# Checking to see if they are both standardized 

customer_pairs = set(zip(customers['customer_city'], customers['customer_state']))
geo_pairs = set(zip(geo['geolocation_city'], geo['geolocation_state']))

missing = customer_pairs - geo_pairs

print(len(missing))
print(missing)
```

```python
# I counted the zips in customer vs geological
total_customer_zips = len(customer_zips)
missing_count = len(missing_zips)

coverage = (total_customer_zips - missing_count) / total_customer_zips

print("Total customer zip prefixes:", total_customer_zips)
print("Missing zip prefixes:", missing_count)
print("Coverage %:", coverage * 100)

# since the coverage percentage is >= 98% then the dataset is good.
```

```python
import re
import unicodedata
import pandas as pd

# Cleaning review_comment_title
# ── Helpers ───────────────────────────────────────────────────────────────────

# Emoji detection: any character outside the Basic Latin + Latin-1 / extended
# Latin ranges that is also a "So / Sm / Sk / Cs / Co" category (symbol / surrogate / private)
_EMOJI_RE = re.compile(
    "["
    "\U0001F300-\U0001FFFF"   # misc symbols, emoticons, transport, etc.
    "\U00002702-\U000027B0"   # dingbats
    "\U000024C2-\U0001F251"   # enclosed chars
    "]+",
    flags=re.UNICODE,
)

# Backtick / accent-like noise that isn't a real diacritic
_NOISE_ACCENT_RE = re.compile(r"[`´]+")

# Date fragments embedded in titles (e.g. "25/07/20", "16/07/2018", "17/05/")
_DATE_RE = re.compile(r"\b\d{1,2}/\d{1,2}(?:/\d{2,4})?\b/?")

# Star globs: any run of * characters (with optional surrounding spaces)
_STARS_RE = re.compile(r"\*+")

# Repeated-character strings of length ≥ 2 that are not actual words
# e.g. "xxx", "aaa", "00", "...." — we'll check these after other cleaning
_REPEATED_CHAR_RE = re.compile(r"^(.)\1+$")

# "Nota" + digit(s) shorthand → keep the text, discard pure digit grade below
_NOTA_RE = re.compile(r"^nota\s*(\d+)\s*$", re.IGNORECASE)

# Titles that are purely numeric (possibly with spaces / punctuation between digits)
_PURE_NUMBER_RE = re.compile(r"^\d[\d\s.,]*$")

# Ratings expressed as large meaningless numbers (1000000, 100, etc.)
_BIG_NUMBER_RATING = re.compile(r"^\d{3,}$")

# Standalone score fractions like "9/10", "10/10"
_SCORE_FRACTION_RE = re.compile(r"^\d{1,2}/\d{1,2}$")

# Titles that are *only* punctuation / symbols / emojis after stripping text
_ONLY_NON_ALPHA_RE = re.compile(r"^[^a-zA-ZÀ-ÿ0-9]+$")

# Garbled / keyboard-smash: short strings with no vowel (e.g. "xx", "ss", "Mm", "ip")
_NO_VOWEL_RE = re.compile(r"^[^aeiouáéíóúàèìòùãõâêîôûäëïöüAEIOUÁÉÍÓÚÀÈÌÒÙÃÕÂÊÎÔÛÄËÏÖÜ]{1,5}$")


def _strip_emojis(text: str) -> str:
    return _EMOJI_RE.sub("", text).strip()


def _count_alpha(text: str) -> int:
    return sum(c.isalpha() for c in text)


def clean_title(raw) -> str | None:
    # ── 0. Null / empty guard ─────────────────────────────────────────────────
    if pd.isna(raw):
        return None
    text = str(raw).strip()
    if not text:
        return None

    # ── 1. Strip emojis (keep surrounding text) ───────────────────────────────
    text = _strip_emojis(text)
    text = text.strip()
    if not text:
        return None

    # ── 2. Remove noise accent characters (` ´) ───────────────────────────────
    text = _NOISE_ACCENT_RE.sub("", text)

    # ── 3. Remove embedded dates (they add no semantic value to a title) ───────
    text = _DATE_RE.sub("", text).strip()

    # ── 4. Star → star-rating label ───────────────────────────────────────────
    #   Count stars BEFORE wiping them so we can produce a label.
    star_match = _STARS_RE.search(text)
    if star_match:
        n_stars = len(star_match.group().replace(" ", ""))
        # Replace the whole star block with a readable label only if ≤ 5 stars
        star_label = f"{min(n_stars, 5)} estrelas" if n_stars <= 5 else ""
        text = _STARS_RE.sub(f" {star_label} ", text).strip()
        # Collapse any double spaces introduced
        text = re.sub(r"\s{2,}", " ", text).strip()

    # ── 5. Strip leading/trailing punctuation (. ! ? , ; : …) ────────────────
    text = text.strip(".,!?;:…\"'")
    text = text.strip()

    # ── 6. Normalize internal whitespace ──────────────────────────────────────
    text = re.sub(r"\s+", " ", text).strip()

    # ── 7. Lowercase ──────────────────────────────────────────────────────────
    text = text.lower()

    # ── 8. Nullify if nothing meaningful remains ──────────────────────────────

    # 8a. Empty after all transforms
    if not text:
        return None

    # 8b. Pure number (single score like "10", "8", "5")
    if _PURE_NUMBER_RE.fullmatch(text):
        return None

    # 8c. Large meaningless number ("1000000", "100")
    if _BIG_NUMBER_RATING.fullmatch(text):
        return None

    # 8d. Score fraction ("9/10", "10/10")
    if _SCORE_FRACTION_RE.fullmatch(text):
        return None

    # 8e. Only punctuation / symbols left (no alphanumeric)
    if _ONLY_NON_ALPHA_RE.fullmatch(text):
        return None

    # 8f. Repeated single character ("xxx", "aaa", "....")
    if _REPEATED_CHAR_RE.fullmatch(text):
        return None

    # 8g. Emoticon shorthand with zero letters (":)", ":D", "??")
    if _count_alpha(text) == 0:
        return None

    # 8h. Very short (≤ 4 chars) with no vowel → keyboard smash / initials noise
    if _NO_VOWEL_RE.fullmatch(text):
        return None

    # 8i. Standalone "nota N" → null (already covered by pure-number if N removed;
    #     handle the composite case explicitly)
    if _NOTA_RE.fullmatch(text):
        return None

    # 8j. Lone star-count label with nothing else ("1 estrelas" is okay, but
    #     if the whole title was just stars and there's no other word, nullify)
    only_star_label = re.fullmatch(r"\d estrelas", text)
    if only_star_label:
        # A pure star rating with no other context is not a meaningful title
        return None

    return text


# ── Apply ─────────────────────────────────────────────────────────────────────
idx = reviews_df.columns.get_loc("review_comment_title")
reviews_df.insert(
    loc=idx + 1,
    column="clean_title",
    value=reviews_df["review_comment_title"].apply(clean_title),
)

# # ── Diagnostics ───────────────────────────────────────────────────────────────
# original_non_null = reviews_df["review_comment_title"].notna().sum()
# clean_non_null    = reviews_df["clean_title"].notna().sum()
# nullified         = original_non_null - clean_non_null

# print(f"Original non-null titles : {original_non_null:,}")
# print(f"After cleaning non-null  : {clean_non_null:,}")
# print(f"Nullified (noise removed): {nullified:,}")
# print(f"Unique cleaned titles    : {reviews_df['clean_title'].dropna().nunique():,}")

# print("\n── Short survivors (len ≤ 5) ────────────────────────────────────────────")
# short_survivors = reviews_df[
#     reviews_df["clean_title"].notna() & (reviews_df["clean_title"].str.len() <= 5)
# ]["clean_title"].value_counts()
# print(short_survivors.to_string())

# print("\n── Sample of newly nullified titles (what was discarded) ────────────────")
# nullified_mask = reviews_df["review_comment_title"].notna() & reviews_df["clean_title"].isna()
# print(
#     reviews_df.loc[nullified_mask, "review_comment_title"]
#     .value_counts()
#     .head(30)
#     .to_string()
# )

# print("\n── Top 20 cleaned titles ────────────────────────────────────────────────")
# print(reviews_df["clean_title"].value_counts().head(20).to_string())

# print("\n── Bottom 20 (rarest) cleaned titles ────────────────────────────────────")
# print(reviews_df["clean_title"].value_counts().tail(20).to_string())
reviews_df = reviews_df.drop(columns=['review_comment_title']).rename(columns={'clean_title': 'review_comment_title'})

```

```python
# Cleaning review comment message

def clean_message(text):
    if pd.isna(text):
        return None

    # Force string
    text = str(text)

    # Decode literal \r\n escape sequences (stored as text, not real newlines)
    text = text.replace('\\r\\n', ' ').replace('\\n', ' ').replace('\\r', ' ')

    # Replace real newlines/tabs with a space
    text = re.sub(r'[\r\n\t]+', ' ', text)

    # Strip leading/trailing whitespace
    text = text.strip()
    if not text:
        return None

    # Normalize internal whitespace
    text = re.sub(r'\s+', ' ', text)

    # Collapse repeated punctuation (e.g. "!!!" → "!", "???" → "?")
    text = re.sub(r'([!?])\1+', r'\1', text)
    text = re.sub(r'\.{2,}', '...', text)  # normalize ellipses

    # Strip trailing punctuation noise (lone "." or "," at the end)
    text = re.sub(r'[\s,;]+$', '', text).strip()

    # Nullify if nothing meaningful remains (pure punctuation / symbols)
    if re.fullmatch(r'[^\w\s]*', text):
        return None

    # Nullify if purely numeric
    if re.fullmatch(r'[\d\s.,]+', text):
        return None

    # Lowercase
    text = text.lower()

    return text

idx = reviews_df.columns.get_loc('review_comment_message')
reviews_df.insert(
    loc=idx + 1,
    column='clean_message',
    value=reviews_df['review_comment_message'].apply(clean_message)
)

reviews_df = reviews_df.drop(columns='review_comment_message').rename(columns={
    'clean_message':'review_comment_message'
})

# print("Original non-null messages:", reviews_df['review_comment_message'].notna().sum())
# print("After cleaning non-null   :", reviews_df['clean_message'].notna().sum())
# print("Nullified                 :", reviews_df['review_comment_message'].notna().sum() - reviews_df['clean_message'].notna().sum())
```
